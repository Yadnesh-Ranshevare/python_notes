# how it will work?
- text = `"hi my name is abc"`
- first break the sentence into the word:
  - word 1 = `"hi"`
  - word 2 = `"my"`
  - word 3 = `"name"`
  - word 4 = `"is"`
  - word 5 = `"abc"`
- at the time of traning model:
  - intput = `"hi"` -> output = `"my"`
  - input = `"hi my"` -> output = `"name"`
  - input = `"hi my name"` -> output = `"is"`
  - input = `"hi my name is"` -> output = `"abc"`

# computer dosent understand english language
text `"hi my name is abc"` is an englidh sentence, therefore we need to conver them into a numbers i.e, vectors

- word 1 = `"hi"` -> 1
- word 2 = `"my"` -> 2
- word 3 = `"name"` -> 3
- word 4 = `"is"` -> 4
- word 5 = `"abc"` -> 5

now inside model
- input = `[1]` -> output `2`
- input = `[1 2]` -> output `3`
- input = `[1 2 3]` -> output `4`
- input = `[1 2 3 4]` -> output `5`

### Convert the text into dataset i.e, inout output format
- for now we have `[93, 1, 13]`
- we want:
  - `[93, 1]` -> input `93` ; output = `1`
  - `[93, 1, 13]` -> input `[93, 1]` ; output = `13`

````python
df = []

for sen in tokenize_sentence: # looping over the complete sentence list
  for i in range(1, len(sen)):   # looping over each token from individual sentence
    df.append(sen[:i+1])
````

### Train test split

````python
y = padded_df[:,-1]
y[0]
````

### One hot encoding on y
As this is the multiclass classification problem will be using sigmoid activation function, but sigmoid only give output between 0 and 1

Therefor:
- lets suppose we have 3 classes
- `class A`, `class B`, `class C`
- in one hot encoded format:
  - `class A` = `[1, 0, 0]`
  - `class B` = `[0, 1, 0]`
  - `class C` = `[0, 0, 1]`

inside model
- our neural network will have 3 output nodes:
  - node 1 -> `class A`
  - node 2 -> `class C`
  - node 3 -> `class B`
- since we have sigmoid activation it will provide a number between 0 to 1
- lets suppose:
  - node 1 -> `0.25`
  - node 2 -> `0.78`
  - node 3 -> `0.43`
- output = `[0.25, 0.78, 0.43]`
- largest output = `0.78` -> node 2
- predicted class = `class B`

````python
len(tokenizer.word_index) # Total number of classes
````

````python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=283)
````

````python
y[0]
````

````python
y.shape # Must have 283 rows as total classes are 283
````

### Build the model
Architecture
```
input -> embedding layer -> LSTM -> dense layer -> Output
```
embedding layer:
- for each input token for each vector it will generate its embedding
- input vector = [0, 1, 0] -> 1x3 vector
- embedding vector = [[0, 2], [2, 1], [3, 5]] -> 1x3x2 vector

LSTM
- input = [[0, 2], [2, 1], [3, 5]] -> embedding vector
- timestep 1 = [0, 2] -> $h_1$
- timestep 2 = [2, 1] -> $h_2$
- timestep 3 = [3, 5] -> $h_3$

Dense layer:
- provide output between 0 to 1
- output = [0.1, 0.5, 0.8]
- predicted class = [0, 0, 1]


````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
````

````python
model = Sequential()
model.add(Embedding(283, 100, input_shape=(56,))) # input = 1x56 ; output = 1x56x100
model.add(LSTM(150))
model.add(Dense(283, activation='softmax'))
````

````python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
````

````python
model.summary()
````

### Traing the model

````python
model.fit(X,y,epochs=20)
````
