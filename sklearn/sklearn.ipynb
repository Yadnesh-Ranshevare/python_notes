# Classifier


```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x,y = load_breast_cancer(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
scalar = StandardScaler()

x_train_scale = scalar.fit_transform(x_train)
x_test_scale = scalar.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

cls = KNeighborsClassifier()
```

```python
cls.fit(x_train_scale, y_train)
```

```python
cls.score(x_test_scale, y_test)
```

```python
input = x_test[2]
cls.predict([input])
```

```python
cls.predict_proba([input])
```

# Regressor

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x,y = fetch_california_housing(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
scalar = StandardScaler()

x_train_scale = scalar.fit_transform(x_train)
x_test_scale = scalar.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(x_train_scale, y_train)
```

```python
reg.score(x_test_scale, y_test)   # R^2 score
```

```python
input = x_test_scale[0]

reg.predict([input])
```

# Metrix

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

x,y = load_breast_cancer(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

scalar = StandardScaler()
cls = KNeighborsClassifier()

x_train_scale = scalar.fit_transform(x_train)
x_test_scale = scalar.transform(x_test)
```

```python
cls.fit(x_train_scale, y_train)
```

```python
cls.score(x_test_scale, y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
y_pre = cls.predict(x_test_scale)
```

```python
accuracy_score(y_test, y_pre)
```

```python
precision_score(y_test, y_pre)
```

```python
recall_score(y_test, y_pre)
```

```python
f1_score(y_test, y_pre)
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

scalar = StandardScaler()
reg = LinearRegression()

x,y = fetch_california_housing(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

x_train_scale = scalar.fit_transform(x_train)
x_test_scale = scalar.transform(x_test)

reg.fit(x_train_scale, y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
y_pred = reg.predict(x_test_scale)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
root_mean_squared_error(y_test, y_pred)
```
